# 2.0 Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 3.0 Getting data

```python
#load default data as a dataframe
tips = sns.load_dataset('tips')
```

# 5.0 Plotting regressions

## 5.1 Linear regression

```python
sns.lmplot(x='total_bill',y='tip',data=tips,palette='Set1')
plt.show()
```

## 5.2 Grouping factors on same plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='day',palette='Set1')
plt.show()
```

## 5.3 Factors on different subplots

### 5.3.1 Plotting into columns

```python
#grouping on category or "factors" and subplotting into columns
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='Set1')
plt.show()
```

### 5.3.2 Plotting into rows

```python
#grouping on category or "factors" and subplotting into columns
sns.lmplot(x='total_bill',y='tip',data=tips,row='sex',hue='day',palette='Set1')
plt.show()
```

## 5.4 Residuals

Notice the expansion as the total bill increases

```python
#how much the data misses the regression line
sns.residplot(x='total_bill',y='tip',data=tips,color='indianred')
plt.show()
```

## 5.5 Higher order regressions

### 5.5.1 Scatter plots

```python
plt.scatter(tips['total_bill'], tips['tip'], label='data', color='red', marker='x')
plt.show()
```

### 5.5.2 First order (linear regression)

```python
sns.regplot(x='total_bill', y='tip', data=tips, label='order 1', color='blue', scatter=None,order=1)
plt.show()
```

### 5.5.3 Second order (polynomial regression)

```python
sns.regplot(x='total_bill', y='tip', data=tips, label='order 2', color='green', scatter=None,order=2)
plt.show()
```

### 5.5.4 Adding all three together

```python
#adding data points to plot
plt.scatter(tips['total_bill'], tips['tip'], label='data', color='red', marker='x')

#adding first order regression to plot
sns.regplot(x='total_bill', y='tip', data=tips, label='order 1', color='blue', scatter=None,order=1)

#adding second order regression to plot
sns.regplot(x='total_bill', y='tip', data=tips, label='order 2', color='green', scatter=None,order=2)

# Add a legend and display the plot
plt.legend(loc='upper left')
plt.show()
```

# 6 Visualizing univariate distributions

## 6.1 Strip plots

### 6.1.1 Univariate

```python
sns.stripplot(y='tip',data=tips)
plt.show()
```

### 6.1.2 Grouped strip plot

```python
#Generate strip plot
sns.stripplot(x='day',y='tip',data=tips)

#label the y-axis and show plot
plt.ylabel('Tip ($)')
plt.show()
```

### 6.1.3 Grouped strip plot with jitter to see frequency

```python
#Generate strip plot
sns.stripplot(x='day',y='tip',data=tips,jitter=True)

#label the y-axis and show plot
plt.ylabel('Tip ($)')
plt.show()
```

## 6.2 Swarm plot

### 6.2.1 Basic swarm plot

```python
#Generate plot
sns.swarmplot(x='day',y='tip',data=tips)

#label and show
plt.ylabel('Tips($)')
plt.show()
```

### 6.2.2 Grouping within a swarm plot

Use 'hue' to group the dataset

```python
#Generate plot
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')

#label and show
plt.ylabel('Tips($)')
plt.show()
```

### 6.2.3 Changing orientation

Remember to switch x & y

Use 'orient' to change the orientation

```python
#Generate plot
sns.swarmplot(x='tip',y='day',data=tips,hue='sex',orient='h')

#label and show
plt.ylabel('Tips($)')
plt.show()
```

## 6.3 Box plots

```python
#Generate plot
sns.boxplot(x='day',y='tip',data=tips)

#label and show
plt.ylabel('Tips ($)')
plt.show()
```

## 6.4 Violin plot

```python
#Generate plot
sns.violinplot(x='day',y='tip',data=tips)

#label and show
plt.ylabel('Tips ($)')
plt.show()
```

## 6.5 Recap on subplots

Remember when you need to change the label for subplots you need to do it for every subplot. It is also often useful to tighten the layout with plt.tight_layout()

```python
#Create and select first subplot
plt.subplot(1,2,1)

#generate plot for first subplot
sns.boxplot(x='day',y='tip',data=tips)

#add label
plt.ylabel('Tips ($)')

#Create and select second subplot
plt.subplot(1,2,2)

#generate plot for first subplot
sns.violinplot(x='day',y='tip',data=tips)

#label and tighten layout
plt.ylabel('Tips ($)')
plt.tight_layout()

#show plot
plt.show()
```

## 6.6 Combining plots

```python
#inner=None disables inner box plot
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgray')

#size=4 decreases default
sns.stripplot(x='day',y='tip',data=tips,jitter=True,size=4)

#label and show
plt.ylabel('Tips ($)')
plt.show()
```
